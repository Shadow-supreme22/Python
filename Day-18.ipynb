# Question 1:
write a program to find the given string in the root directory that contains a text file
1. In a recursive manner (case in sensitive )
2. only in the root directory
use command line arguments.
program_name -ri string root_dir (recursive)
program_name -si string root_dir (root director )
The output which file and line contains the string

```python
import os
import re
import sys
import argparse

class St_search :
    def __init__(self, string2, path1,i=None):
        self.path1= path1
        self.string1 = string2
        self.i=i
        if self.i:
            string2 = string2.lower()
            self.string2= re.compile(string2)
def txt_search(self):
    file_number = 0
    files = [f for f in os.listdir(self.path1) if os.path.isfile(self.path1+"/"+f)]
    for file in files:
        file_t = open(self.path1+"/"+file)
        file_text= file_t.read()
        if self.i:
            file_text=file_text.lower()
        file_t.close()
        if re.search(self.string2, file_text):
            print("The text “+self.string1+” found in ", file)
        file_number=file_number+1
        print("total files are ",file_number)
def s_search_m(self):
    files = [f for f in os.listdir(self.path1) if os.path.isfile(self.path1+"/"+f)]
    file_number = 0
    for file in files:
        file_t = open(self.path1+"/"+file)
        line_number=1
        flag_file = 0
    for line1 in file_t:
        if self.i:
            line1 = line1.lower()
            if re.search(self.string2, line1):
                flag_file= 1
        print ("The text "+self.string1+" found in ", file, " at line number ",line_number)
        line_number=line_number+1
        if flag_file == 1:
            file_number=file_number+1
            flag_file=0
            file_t.close() 
        print("total files are ",file_number)
def s_search_r(self):
    file_number = 0
    for root, dir, files in os.walk(self.path1, topdown = True):
        files = [f for f in files if os.path.isfile(root+"/"+f)]
        for file in files:
            file= root+"/"+file
            file_t = open(file)
            line_number=1
            flag_file = 0
            for line1 in file_t:
                if self.i:
                    line1=line1.lower()
                    if re.search(self.string2, line1):
                        flag_file= 1
            print("The text “+self.string1+” found in ", file," at line number",line_number)
            line_number=line_number+1
            if flag_file == 1:
                file_number=file_number+1
                flag_file=0
                file_t.close() 
        print("total files are ",file_number)
def main():
    parser = argparse.ArgumentParser()
    parser.add_argument('-si', nargs = 2)
    parser.add_argument('-ri', nargs = 2)
    args = parser.parse_args()
        try:
            if args.si:
                if args.si[1]:
                    dir = args.si[1]
                    obj1 = St_search(args.si[0],dir,i=1)
                    obj1.txt_search()
            elif args.ri:
                if args.ri[1]:
                    dir = args.ri[1]
                    obj1 = St_search(args.ri[0],dir,i=1)
                    obj1.s_search_r()
            elif args.mi:
                dir = args.mi[1]
                obj1 = St_search(args.mi[0],dir,i=1)
                obj1.s_search_m()
        except Exception as e:
            print(e)
main()
```
